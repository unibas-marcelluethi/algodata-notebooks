# Logarithmen

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Der Logarithmus ist die Umkehrfunktion der Exponentialfunktion.

```python
x = np.linspace(0.01, 50, 500)
#plt.plot(x, np.log2(2**x), label=r'$\log_{2}(2^x)$')
plt.plot(x, 2**(np.log2(x)), label=r'$2^{\log_{2}(x)}$')
plt.legend()
```

Der Logarithmus wächst (ab einem gewissen Wert) sehr langsam.

```python
x = np.linspace(0.01, 100000, 1000)
plt.plot(x, np.log2(x), ls="-", label=r'$\log_2x$')
plt.plot(x, np.log(x), ls="-", label=r'$\log_ex$')
plt.plot(x, np.log10(x), ls="-", label=r'$\log_{10}x$')
plt.legend()
```

Wir können die Basisumrechnungsformel $\log_b x = \log_a x/ \log_a b$ verwenden, um Logarithmen zur einer beliebigen Basis zu berechnen:

```python
basis = 8
plt.plot(x, np.log2(x)/np.log2(basis), ls="-", label=r'$\log_{%s}x$' % basis)
plt.legend()
```

Beobachtung: Ersetzen wir die Basis $a$ durch Basis $b$, verändert sich der Wert um einen konstanten Faktor.
